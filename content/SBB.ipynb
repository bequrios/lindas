```python
df = await query("""

PREFIX schema: <http://schema.org/>

SELECT DISTINCT ?name ?geometry ?ele WHERE {
    
    ?relation <http://vocab.gtfs.org/terms#stop> ?stop.
    ?stop schema:name ?name;
        <http://www.opengis.net/ont/geosparql#hasGeometry>/<http://www.opengis.net/ont/geosparql#asWKT> ?geometry;
        schema:elevation ?ele.
    
    {
        SELECT ?relation (COUNT(?id) as ?number) WHERE {

            { 

                SELECT DISTINCT * WHERE {

                    ?relation a <https://lod.opentransportdata.swiss/vocab/Relation>;
                        <https://lod.opentransportdata.swiss/vocab/zone> <https://lod.opentransportdata.swiss/zone/libero/libero-billett-libero/101>;
                        <https://lod.opentransportdata.swiss/vocab/zone>/schema:identifier ?id.

                }
            }

        } GROUP BY ?relation
    }
    FILTER(?number = 1)
    
}
    
""")

display_result(df)
```

```python
df = gpd.GeoDataFrame(df)

df["geometry"] = gpd.GeoSeries.from_wkt(df["geometry"], crs="EPSG:4326")

df['lat'] = df.geometry.y
df['long'] = df.geometry.x

sw = df.bounds[["miny", "minx"]].min().values.tolist()
ne = df.bounds[["maxy", "maxx"]].max().values.tolist()

display(df.head())
```

```python
# create the map and center it around the mean coordinates of the stations
m = folium.Map(location=[df['lat'].mean(),df['long'].mean()], tiles="openstreetmap", zoom_start=8)

# add marker one by one on the map
for i in range(0,len(df)):
    folium.Marker(
        location=[df.iloc[i]['lat'], df.iloc[i]['long']],
        popup=folium.Popup(folium.Html('<h3>' + df.iloc[i]['name'] + '</h3>', script=True), max_width=500),
        icon=folium.Icon(icon='train', prefix='fa')
   ).add_to(m)

m.fit_bounds([sw, ne])
# show the map
m
```
